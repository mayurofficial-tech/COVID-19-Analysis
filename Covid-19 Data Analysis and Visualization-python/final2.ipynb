```python
import matplotlib.pyplot as plt
import numpy as np

# Data
continents = ["Africa", "Asia", "Europe", "North America", "Oceania", "South America"]
cases = [14616886098, 252167317226, 236756684151, 127073670231, 11598053992, 73484570403]
deaths = [306955815, 1790404880, 2361106039, 1880636077, 23933802, 1646924372]
survive = [14309930283, 250376912346, 234395578112, 125193034154, 11574120190, 71837646031]

# Set up bar width and position
x = np.arange(len(continents))
width = 0.28

# Create figure
plt.figure(figsize=(12,7))

# Plot grouped bar chart
bars1 = plt.bar(x - width, cases, width, label="Cases", color="#3498db")
bars2 = plt.bar(x, deaths, width, label="Deaths", color="#e74c3c")
bars3 = plt.bar(x + width, survive, width, label="Survive", color="#2ecc71")

# Add labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height, f"{height/1e9:.1f}B", 
                 ha='center', va='bottom', fontsize=9, rotation=90, fontweight="bold")

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# Titles and formatting
plt.title("COVID-19 Cases, Deaths, and Survivors by Continent", fontsize=15, fontweight="bold")
plt.ylabel("Number of People (Billions)", fontsize=12)
plt.xticks(x, continents, fontsize=11)
plt.legend(fontsize=11)
plt.savefig("C://Users//MAYUR MAKVANA//OneDrive//Desktop//project chart//continent2.png")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
continents = ["Africa", "Asia", "Europe", "North America", "Oceania", "South America"]
population = [2387392859466, 7850018689974, 1251166465980, 1004941801818, 75395130318, 731231120646]
cases = [14616886098, 252167317226, 236756684151, 127073670231, 11598053992, 73484570403]

# Calculate cases per 100 people (cases/population ratio)
cases_perc = [c/p * 100 for c, p in zip(cases, population)]

# Create figure
plt.figure(figsize=(10,6))

# Bar chart of percentage of cases relative to population
bars = plt.bar(continents, cases_perc, color="#3498db", width=0.6)

# Add value labels on bars
for bar, perc in zip(bars, cases_perc):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{perc:.2f}%", 
             ha='center', va='bottom', fontsize=10, fontweight="bold")

# Titles and formatting
plt.title("COVID-19 Cases as % of Population by Continent", fontsize=15, fontweight="bold")
plt.ylabel("Cases (% of Population)", fontsize=12)
plt.xlabel("Continent", fontsize=12)
plt.savefig("C://Users//MAYUR MAKVANA//OneDrive//Desktop//project chart//continent3.png")
plt.tight_layout()
plt.show()
```
